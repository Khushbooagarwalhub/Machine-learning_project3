## Preliminaries
Test the best model with photos from google images

```python
from tensorflow.keras.applications import inception_resnet_v2
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image
import pandas as pd
import numpy as np
import requests
import joblib
import cv2
```

## Import class_indices.csv

```python
file = "../Output/image_features/ten/class_indices.csv"
classes_df = pd.read_csv(file); del classes_df['Unnamed: 0']
```

## Prediction
The process is two-fold:
- Feed the image into a pre-trained model to extract predicted features
- Feed the features vector into a classifier to idenity action class

Pre-trained model

```python
inception = inception_resnet_v2.InceptionResNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

Classifier

```python
model_path = "../Output/models/ten/LR.sav"
classifier = joblib.load(model_path)
```

Some image urls

```python
urls = ["https://s3.envato.com/files/255225318/_MG_5041.jpg", 
         "https://previews.123rf.com/images/sheftsoff/sheftsoff1708/sheftsoff170805882/84081992-girl-sings-and-playing-guitar-musician-with-acoustic-guitar.jpg",
         "http://cdn.akc.org/content/hero/leash_walking_hero.jpg"
        ]
```

Load image and classify

```python
for url in urls:
    response = requests.get(url)
    img = np.uint8(Image.open(BytesIO(response.content)))
    img = img/255
    img = cv2.resize(img, (224, 224))
    plt.imshow(img)
    plt.show()
    
    img_arr = np.expand_dims(img, axis=0)
    features = inception.predict(img_arr)
    features_shaped = np.reshape(features, (1, 5*5*1536))
    
    prediction_id = classifier.predict(features_shaped)
    prediction_class = classes_df['Class'].loc[classes_df['ID'] == int(prediction_id)].values[0]
    print(prediction_class)
```

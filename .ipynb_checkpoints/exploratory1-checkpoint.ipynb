```python
import matplotlib.pyplot as plt
import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(11, 4)})
```

```python
#pltting the time series data
data['acc_x'].plot(linewidth=0.5);
```

```python
data_running = data[data['activity'] =='Jogging']
data_running
#taking the first 200 records 
data_running_filter = data_running[:200]
data_running_filter
```

```python
fig, ax = plt.subplots(figsize=(15, 7))

# Add the x-axis and the y-axis to the plot
ax.plot(data_running_filter['timestamp'],
        data_running_filter['acc_x'],
        color='purple')

ax.set(xlabel="timestamp",
       ylabel="X acceleration",
       title="jogging")


fig, ax = plt.subplots(figsize=(15, 7))

ax.plot(data_running_filter['timestamp'],
        data_running_filter['acc_y'],
        color='purple')

ax.set(xlabel="timestamp",
       ylabel="y acceleration",
       title="jogging")



fig, ax = plt.subplots(figsize=(15, 7))

ax.plot(data_running_filter['timestamp'],
        data_running_filter['acc_z'],
        color='purple')

ax.set(xlabel="timestamp",
       ylabel="z acceleration",
       title="jogging")
```

```python
data_walking = data[data['activity'] =='Walking']
data_walking
#taking the first 200 records 
data_walking_filter = data_walking[:200]
data_walking_filter
```

```python
fig, ax = plt.subplots(figsize=(15, 7))

# Add the x-axis and the y-axis to the plot
ax.plot(data_walking_filter['timestamp'],
        data_walking_filter['acc_x'],
        color='red')

ax.set(xlabel="timestamp",
       ylabel="X acceleration",
       title="walking")


fig, ax = plt.subplots(figsize=(15, 7))

ax.plot(data_walking_filter['timestamp'],
        data_walking_filter['acc_y'],
        color='red')

ax.set(xlabel="timestamp",
       ylabel="y acceleration",
       title="walking")



fig, ax = plt.subplots(figsize=(15, 7))

ax.plot(data_walking_filter['timestamp'],
        data_walking_filter['acc_z'],
        color='red')

ax.set(xlabel="timestamp",
       ylabel="z acceleration",
       title="walking")
```

```python
data_standing = data[data['activity'] =='Standing']
data_standing
#taking the first 200 records 
data_standing_filter = data_standing[:200]
data_standing_filter
```

```python
fig, ax = plt.subplots(figsize=(15, 7))

# Add the x-axis and the y-axis to the plot
ax.plot(data_standing_filter['timestamp'],
        data_standing_filter['acc_x'],
        color='blue')

ax.set(xlabel="timestamp",
       ylabel="X acceleration",
       title="standing")


fig, ax = plt.subplots(figsize=(15, 7))

ax.plot(data_standing_filter['timestamp'],
        data_standing_filter['acc_y'],
        color='blue')

ax.set(xlabel="timestamp",
       ylabel="y acceleration",
       title="standing")



fig, ax = plt.subplots(figsize=(15, 7))

ax.plot(data_standing_filter['timestamp'],
        data_standing_filter['acc_z'],
        color='blue')

ax.set(xlabel="timestamp",
       ylabel="z acceleration",
       title="standing")
```
